```julia
function meuler(f,x0,listas)#definimos una fución con inputs:f=f(t,x),la condición inicial x0 y la lista que contiene
#la partición del tiempo
listx=zeros(length(listas))#la lista de la partición en x debe ser del mismo tamaño que lista del tiempo
h=(listas[end]-listas[1])/(length(listas)-1)# calculamos el tamaño de la diferencia finita h
listx[1]=x0#el primer elemento de la lista en x es la condición inicial
t=listas[1]#asignamos el primer elemento de la partición en t a la lista de tiempo
x=x0#iniciamos a x con la condición inicial
    for i in 2:length(listas)# este ciclo for es para implementar el método de euler desde el segundo elemento hasta el 
        #último elemento de la partición de tiempo
        x=x+(h*f(t,x))#se implementa el método de euler
         listx[i]=x#se asigna x a cada elemento de la lista en x
        t=listas[i]# se asigna como nuevo valor de t al siguiente punto de tiempo
end
return listx  # se regresa la partición de x
end
```

Ejercicio 2

El sistema físico descrito por 
$$ \ddot{x} =-kx,    x(0)=a ,    x'(0)=b,  \omega^{2}=k$$

es el oscilador armónico. Es decir u  sistema que al salir del equilibrio tiende a regresar a él.

Hacemos la sustitución x =$e^{\lambda t}$ 

$\Rightarrow \lambda^{2}=-\omega^{2}$

$\Rightarrow \lambda=\pm i\omega$

$\Rightarrow$  x = $e^{i\omega t}+e^{-i\omega t}$

Tomando la parte real de x y por la linealidad de la ecuación, podemos tomar  las soluciones:

x=  $c_{1} cos(\omega t)$+ $c_{2} sin(\omega t)$

De la condición inicial $x(0)=a$ 

$\Rightarrow c_{1}=a$

Ahora calculemos


$\dfrac{dx}{dt}=-\omega c_{1} sin(\omega t)+  \omega c_{2} cos(\omega t)$

Sustituyendo $x'(0)=b$

$\Rightarrow  \omega c_{2} = b$

Finalmente la solución de la ecuación junto con las condiciones iniciales es:

x=  $a  cos(\omega t)$+ $\dfrac{b}{\omega} sin(\omega t)$

Hacemos  $u = \dot{x}(t)  \Rightarrow$ Del sistema original observamos $\dot{u} =\ddot{x}=-kx$. 

Entonces definiendo el vector $\bar{Y}$
 
$\bar{Y}=(x,u) \Rightarrow \dfrac{d\bar{Y}}{dt}=(\dot{x},\dot{u})=(\dot{x},-kx)$

Es decir si tenemos un sistema

 $  \dfrac{d\bar{Y}}{dt}=\bar{F}(\bar{Y})$,
 
 $\bar{F}(\bar{Y})=(F(x,u),G(x,u))$ con la condición inicial  $ \bar{Y_{0}}=(x_{0},u_{0})$
 Entonces podemos aproximar por el método de Euler:
 
 
 $(x_{k+1},u_{k+1})=(x_{k},y_{k})+ h\bar{F}(x_{k},u_{k})$,   con h el tamaño de la partición uniforme de t

Ejericio 4

Ahora se resuelve el sistema $  \dfrac{d\bar{Y}}{dt}=\vec f(\vec x,t)=(u,-kx)$, con k=2.5, correspondiente al oscilador armónico,  $\ddot{x} =-kx$,
con condición inicial  x(0)=0, u=x'(0)=1.

La solución que se espera obtener es:   $0.632456sin(1.58114t)$

```julia
listT=linspace(0,8,101)#creamos una partición del intervalo de t (0,8)
```

```julia
function F(x,u)#se define la función que será la primera entrada del vector F
return u
end
```

```julia
function G(x,u)#se define la función que será la segunda entrada del vector F
return -2.5*x
end
```

```julia
function meulervec(F,G,x0,u0,listas)#se implemeta el método del Euler para sistemas de 2 dimensiones, con input, 
#las funciones componentes del sistema F, G las condiciones iniciales, x(0), u(0=x'(0) y la lista con la partición de
# del tiempo t
listx=zeros(length(listas))# se crean las listas de  x y u  con ceros donde se guardarán los datos, del mismo tamaño
# que la lista de t
listu=zeros(length(listas))
    h=(listas[end]-listas[1])/(length(listas)-1)#se calcula el tamaño del paso el t
listx[1]=x0 #se inicia los valores de x y u de las listas con las condiciones dadas en el input
listu[1]=u0
t=listas[1]#se asigna a t el primer elemento de la lista de t
x=x0#se asigna los valores de x y u con los dados en el input
u=u0
for i in 2:length(listas)#en este for se calcula el método de Euler para cada componente del vector asignado valores desde
        #desde el elemento 2 hasta el último de la lista de t
        x=x+(h*F(x,u))
        listx[i]=x #se asigna cada elemento del método de euler a la lista x
        u=u+(h*G(x,u))
        listu[i]=u#se asigna cada elemento del método de euler a la lista u
        t=listas[i]
end
return (listx,listu)  #se regresa la lista de x y u
end
```

```julia
(listX,listU)=meulervec(F,G,0,1,listT)#asignamos las listas de x y u respectivamente
```

```julia
t=listT
Z=0.632456sin.(1.58114t)
```

```julia
xlabel("t")
ylabel("x")
plot(listT,listX,".")#ahora se grafican las solucion numérica en  X y observamos que se obtiene una función sinusoidal
plot(listT,Z)#en azul los pasos de la solución numérica y en naranja la solución exacta
```

```julia
xlabel("t")
ylabel("u=x'")
plot(listT,listU)#ahora se grafican las solucion  U y observamos que se obtiene una función sinusoidal
```
