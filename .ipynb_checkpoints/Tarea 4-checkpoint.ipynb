```julia
function polinomio(init)
    
list=zeros(100);
f(x)=x^3-x^2-x-1;
f´(x)=3x^2-2x-1;
x=init
for i in 1:100
   x=x-f(x)/f´(x)
    list[i]=x;
end
    return list
end;
```

```julia
list1=polinomio(2.0);
list2=polinomio(1.1);
list3=polinomio(4.0);
list4=polinomio(3.5);
```

```julia

plot(real(list1),"o")
plot(real(list2),"o")
plot(real(list3),"o")
plot(real(list4),"o")
xlabel("Número de la iteración")
ylim(1,5)
xlim(-1,30)
show();
```

```julia
S=list1[end]
x=(list1-S)/S
plot(x,"o")
```

```julia
function biseccion(a,b)
    list5=zeros(100)
    p = (a+b)/2
    f(x) = x^3-x^2-x-1
    for i in 1:100
        if  f(a)*f(p) > 0
            a = p
        list5[i]=p;
        end
        
        if f(b)*f(p) > 0
            b = p
        list5[i]=p;
        end
        p = (a+b)/2
    end
    return list5
end;
```

```julia
biseccion(1,2)
```

```julia
list5=biseccion(1,2);
```

```julia
plot(real(list5),"o");
xlabel("Número de la iteración")
ylim(1,3)
xlim(-1,30)
title("Convergencia método de bisección")
show();
```

```julia
S=list5[end]
x=(list4-S)/S
plot(x,"o")
```
