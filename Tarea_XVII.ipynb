```julia
using PyPlot
```

```julia
listlyap = []#se crea una lista para guardar los valores del exp. de liapunov
listx=[]#se crea una lista para guardar los valores de la variable x
j = 0 #iniciamos el contador
for r in 3:0.0001:4# en el for se obtiene los valores del exp de liapunov para una partición en r
    xn = rand()#damos condiciones iniciales aleatorias para x
    ly = 0#iniciamos la suma en 0
    j = j + 1# cambiamos el contador
    for i in 1:10000#en este for aplicamos el valor de xn con el mapeo logístico
        
        xn = r*xn*(1 - xn)
       
        if i>300
            ly += log(abs(r-2*r*xn))#calculamos el valor del exponente de lyapunov
       
        end
        
        xn = r*xn*(1 - xn)
    end
     push!(listx,xn)#se guarda el valor de xn en listx
    ly = ly/10000#divinimos entre n
    push!(listlyap,ly)#se guarda el valor del exp de lapunov

end

r=3:0.0001:4;


plot(r,listx,".")
title("Diagrama de Bifurcaciones")
xlabel("r");ylabel("xn")
```
