```julia
function RK4vec(f,list,x0)#se crea la función que toma como entradas, la función vectorial, lista de tiempo y la condición
    #inicial x(0)
     x = x0#se inicia la variable x en x(0)
     h = list[2]-list[1]#calculamos el tamaño del paso
     listx=[]#se crea la lista  para posteriormente guardar los datos
     push!(listx,x)#se  gurada la los valores de x en listx
    for i in 2:length(list)#en este for se realiza el método de euler para todos los elementos de la listax
    t = i*h#asigna los valores de tiempo
    k1=f(x,t)#se definen todos los coeficientes del método de Runge Kutta de O(4)
    k2=f(x+((h/2)*k1),t+h/2)
    k3=f(x+((h/2)*k2),t+h/2)
    k4=f(x+h*k3,t+h)
    x=x+ (h/6)*(k1+2*k2+2*k3+k4)
    push!(listx,x) #se  gurada la los valores de x en listx
     end
     return listx#se regresa la lista x
end
```

```julia
f(x,t)=[-x[2]-x[2]^2,x[1]-x[2]^3]#se define la nueva función vectorial
```

```julia
using PyPlot
```

```julia
list=-20:0.1:20;
for i in 1:1000#en este for se crean condiciones iniciales aleatorias para ver el comṕortamiento local del sistema
x=(rand())/2
y=(rand())/2
vec1=RK4vec(f,list,[y,x]);
listx3=map(x->x[1],vec1);
listx4=map(x->x[2],vec1);
grid("on")
xlabel("x");ylabel("y"),
plot(listx3,listx4)
end
#observamos que localmente se comporta como un centro
```

 Dada la Matriz A
 $$A=\left(\begin{array}{cc} f_{1}(x,y)  \\f_{2}(x,y) \end{array} \right)$$
 calculamos el Jacobiano como:
$$A=\left(\begin{array}{cc} \frac{\partial f_{1}}{\partial x} &\frac{\partial f_{1}}{\partial y} \\\frac{\partial f_{2}}{\partial x}&\frac{\partial f_{2}}{\partial y} \end{array} \right)$$
 
 Así dada la matriz:
 
 $$F=\left(\begin{array}{cc} -y-y^{3}  \\-x-y^{2} \end{array} \right)$$
 en los puntos 
  el jacobiano es:
   $$JF=\left(\begin{array}{cc} 0&1-3y^{2}  \\-1&-2y \end{array} \right)$$
  
   $$JF(0,0)=\left(\begin{array}{cc} 0 & 1 \\-1&0 \end{array} \right)$$
  
  Entonces: $\Delta=1$, $\tau=0$
  
  Entonces del análisis local (0,0) es unpunto centro.

```julia
g(x,t)=[x[2],x[2]*0.5*(1-x[1]^2)-x[1]]#creamos la funcion vectorial
```

```julia

listt=-10:0.01:10#definimos otra lista de t 
for i in 1:1000# en este for se grafica x y y para 1000 condiciones iniciales
x=4*(2rand()-1)#se genera una condición inicial x0
# aleatoria
y=4*(2rand()-1)#se genera una condición inicial y0
# aleatoria
vec1=RK4vec(g,listt,[x,y]);   
listx=map(x->x[1],vec1);
listy=map(x->x[2],vec1);
grid("on")
xlabel("x");ylabel("y"),
plot(listx,listy);
end
# se observan los ciclos límite de la ecuación de Van der Pol
```
